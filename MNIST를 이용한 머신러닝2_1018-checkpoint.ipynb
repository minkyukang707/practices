```python
from tensorflow.keras.datasets.mnist import load_data
(x_train,y_train),(x_test,y_test) = load_data(path='mnist.npz')

from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val = train_test_split(x_train,y_train, test_size=0.3,
                                              random_state=777)

num_x_train=x_train.shape[0]
num_x_val=x_val.shape[0]
num_x_test=x_test.shape[0]
num_x_train
```

```python
x_train=(x_train.reshape((num_x_train,28*28)))/255
x_val=(x_val.reshape((num_x_val,28*28)))/255
x_test=(x_test.reshape((num_x_test,28*28)))/255

#y_test의 0~9까지의 숫자가 각각 몇개인지 조회하기
type(y_test)
y_test[:10]
```

```python
import pandas as pd 
df=pd.Series(y_test)
df.value_counts().sort_index()
```

```python
from tensorflow.keras.utils import to_categorical
y_train=to_categorical(y_train)
y_val=to_categorical(y_val)
y_test=to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model=Sequential()
model.add(Dense(64, activation='relu', input_shape=(784,)))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['acc'])
history=model.fit(x_train,y_train,
                 epochs=30,
                 batch_size=128,
                 validation_data=(x_val,y_val))
```

```python
import matplotlib.pyplot as plt
his_dict=history.history # loss,acc,val_loss,val_acc
loss=his_dict['loss'] #학습데이터의 손실함수값
val_loss=his_dict['val_loss'] #검증데이터의 손실함수값
epochs=range(1,len(loss)+1) #1~30값
fig=plt.figure(figsize=(10,5))
#훈련 및 검증손실 그리기
ax1=fig.add_subplot(1,2,1) #1행 2열의 1번째 그래프
ax1.plot(epochs,loss, color = 'blue',label='train_loss')
ax1.plot(epochs,val_loss, color = 'orange',label='val_loss')
ax1.set_title('train and val loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('loss')
ax1.legend()

acc=his_dict['acc'] #학습데이터의 정확도
val_acc=his_dict['val_acc'] #검증데이터의 정확도

#훈련 및 검증정확도 그리기
ax2=fig.add_subplot(1,2,2) #1행 2열 2번째 그래프
ax2.plot(epochs, acc,color='blue', label='train_acc')
ax2.plot(epochs, val_acc,color='orange', label='val_acc')
ax2.set_title('train and val acc')
ax2.set_xlabel('epochs')
ax2.set_ylabel('acc')
ax2.legend()
plt.show()
```

```python
#테스트데이터의 손실함수값과 정확도
model.evaluate(x_test,y_test)
```

```python
#혼동행렬 조회하기
results=model.predict(x_test)
results[0]
```

```python
from sklearn.metrics import classification_report, confusion_matrix
#np.argmax(y_test, axis=-1 : y_test의 최대값과 인덱스리턴
#:인덱스가 숫자
import numpy as np
cm=confusion_matrix(np.argmax(y_test,axis=-1),np.argmax(results,axis=-1))
cm
```

```python
#heatmap으로 출력하기
import matplotlib.pyplot as plt 
import seaborn as sns
plt.figure(figsize=(7,7))
sns.heatmap(cm,annot=True, fmt='d', cmap='Blues')
plt.xlabel('predicted label',fontsize=15)
plt.ylabel('true label',fontsize=15)
plt.show()
```

```python
print(df.value_counts().sort_index())
print('\n')
print(classification_report(np.argmax(y_test,axis=-1), np.argmax(results, axis=-1)))
```

# Fashion-MNIST 데이터셋 활용

```python
from tensorflow.keras.datasets.fashion_mnist import load_data
(x_train,y_train), (x_test,y_test) = load_data()
print(x_train.shape, x_test.shape)
print(y_train[:10])
```

```python
class_names=['T-shirt/top','Trouser','Pullover','Dress','Coat','Sandal','Shirt',
            'Sneaker','Bag','Ankle boot']
#임의의 데이터를 이미지로 출력하기
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(777)
sample_size=9

#0~59999사이의 값을 9개를 선택
random_idx=np.random.randint(60000,size=sample_size)
plt.figure(figsize=(5,5))
for i, idx in enumerate(random_idx):
    plt.subplot(3,3,i+1)
    plt.xticks([]) #x축 없음
    plt.yticks([]) #y축 없음
    plt.imshow(x_train[idx],cmap='gray')
    plt.xlabel(class_names[y_train[idx]])
    plt.show()
```

```python
#학습데이터, 테스트 데이터를 정규화하기. 0~1값으로 변경 
#minmax 정규화 : x-min/max-min
x_train[:10]
x_train=x_train/255
x_test=x_test/255
x_train.shape
#레이블을 one-hot인코딩데이터로 변경하기
from tensorflow.keras.utils import to_categorical
y_train=to_categorical(y_train)
y_test=to_categorical(y_test)
print(y_train[:10])
```

```python
#학습데이터의 7:3비율로 검증데이터 분리
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=0.3,
                                            random_state=777)
x_val.shape
```

```python
#모델 구성하기
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten
model1=Sequential()
model1.add(Flatten(input_shape=(28,28))) #2차원 데이터를 1차원형태(28*28)로 변경. 입력층
model1.add(Dense(64,activation='relu'))
model1.add(Dense(32,activation='relu'))
model1.add(Dense(10,activation='softmax'))
model1.summary()
```

```python
model1.compile(optimizer='adam',loss='categorical_crossentropy', metrics=['acc'])
history1=model1.fit(x_train,y_train, epochs=30,
                  batch_size=128, validation_data=(x_val,y_val))
model1.evaluate(x_test,y_test)
```

```python
#혼동행렬
results=model1.predict(x_test) #예측한 값 
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
cm=confusion_matrix(np.argmax(y_test,axis=-1),
                   np.argmax(results,axis=-1))
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(7,7))
sns.heatmap(cm,annot=True, fmt='d',cmap='Blues')
plt.xlabel('predicted label',fontsize=15)
plt.ylabel('true label',fontsize=15)
plt.show()
```

```python
print(classification_report(np.argmax(y_test,axis=-1), np.argmax(results, axis=-1)))
```

```python
#model2 구성하여 학습
#입력층, 128개의 출력을 가지는 은닉층, 64개의 출력을 가지는 은닉층
#32개의 출력을 가지는 은닉층, 출력층으로 이루어진 모델을 구성하기
#모델을 설정하고 학습해서 혼동행렬까지 출력하기
model2=Sequential()
model2.add(Flatten(input_shape=(28,28))) #2차원데이터를 1차원형태(28*28)로 변경.입력층
model2.add(Dense(128,activation='relu'))
model2.add(Dense(64,activation='relu'))
model2.add(Dense(32,activation='relu'))
model2.add(Dense(10,activation='softmax'))
model2.summary()
```

```python
model2.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['acc'])
history2=model2.fit(x_train, y_train, epochs=30,
                   batch_size=128, validation_data=(x_val,y_val))
model2.evaluate(x_test,y_test)
results=model2.predict(x_test) #예측한값
```

```python
cm=confusion_matrix(np.argmax(y_test,axis=-1), np.argmax(results,axis=-1))
cm
```

```python
plt.figure(figsize=(7,7))
sns.heatmap(cm,annot=True, fmt='d',cmap='Blues')
plt.xlabel('predicted label',fontsize=15)
plt.ylabel('true label',fontsize=15)
plt.show()
```

```python
print(classification_report(np.argmax(y_test,axis=-1), np.argmax(results,axis=-1)))
```

```python
#두 모델을 그래프로 비교하기
import numpy as np
import matplotlib.pyplot as plt

def draw_loss_acc(history_1,history_2,epochs):
    his_dict_1=history_1.history
    his_dict_2=history_2.history
    keys=list(his_dict_1.keys()) #['loss','acc','val_loss','val_acc']
    epochs=range(1,epochs)
    fig=plt.figure(figsize=(10,10))
    ax=fig.add_subplot(1,1,1)
    #spines: 그래프의 외곽선
    ax.spines['top'].set_color('none')
    ax.spines['bottom'].set_color('none')
    ax.spines['left'].set_color('none')
    ax.spines['right'].set_color('none')
    ax.tick_params(labelcolor='w',top=False,bottom=False,
                  left=False, right=False)
    for i in range(len(his_dict_1)):
        temp_ax=fig.add_subplot(2,2,i+1)
        temp=keys[i%2]
        val_temp=keys[(i+2)%2+2]
        temp_history=his_dict_1 if i<2 else his_dict_2
        temp_ax.plot(epochs, temp_history[temp][1:],
                    color='blue',label='train_'+temp)
        temp_ax.plot(epochs,temp_history[val_temp][1:],color='orange',
                    label=val_temp)
        if(i==1 or i==3):
            start,end=temp_ax.get_ylim()
            temp_ax.yaxis.set_ticks(np.arange(np.round(start,2),end,0.01))
            temp_ax.legend()
            ax.set_ylabel('loss',size=20)
            ax.set_xlabel('Epochs',size=20)
            plt.tight_layout()
            plt.show()
```

```python
draw_loss_acc(history1,history2,30)
```
