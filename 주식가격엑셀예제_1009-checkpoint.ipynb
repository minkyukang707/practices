# 주식가격, 주식가치 엑셀 예제

```python
import pandas as pd 
```

```python
df1= pd.read_excel('C:/pyy/stockprice.xlsx',engine = 'openpyxl')
df2= pd.read_excel('C:/pyy/stockvaluation.xlsx',engine = 'openpyxl')
```

```python
print(df1)
print(df2)
```

```python
# df1, df2를 열기준으로 연결하기
result1 = pd.concat([df1,df2],axis=1)
print(result1)
print(result1.columns)
```

```python
#df1,df2를 열기준으로 병합하기
#id 컬럼을 기준으로 같은 id를 가진 레코드 병합 
#inner 방식
result2 = pd.merge(df1,df2)
print(result2)
print(result2.columns)
```

```python
#병합시 두개의 데이터프레임의 연결 컬름 설정
#연결 컬럼을 key라고 한다
#outer 방식
result3 = pd.merge(df1,df2, on='id',how='outer')
print(result3)
print(result3.columns)
```

```python
#병합시 사용되는 키의 이름이 다른 경우
#how ='left' : 왼쪽 데이터셋(df1)의 내용 모두 조회
#left_on = 'stock_name',right_on = 'name':키값을 설정
#키로 사용되는 컬럼명이 양쪽이 다르다
result4 = pd.merge(df1,df2, how = 'left', left_on = 'stock_name',right_on = 'name')
print(result4)
print(result4.columns)
print(df1)
```

```python
#df2 데이터셋의 내용을 모두 조회되도록, 양쪽의 키는 df1의 stock_name, df2의 name컬럼
#출력할 최대 열의 개수, ipython의 출력부분 설정
pd.set_option('display.max_columns',10)
#출력할 열의 너비 ipython의 출력부분 설정
pd.set_option('display.max_colwidth',20)
```

```python
result5 = pd.merge(df1,df2, how = 'right',left_on = 'stock_name',right_on = 'name')
print(result5)
print(result5.columns)
```

```python
#df1 데이터 중 금액이 50000 미만인 데이터셋과
#df2 데이터를 id 컬럼으로 병합하기 
price_under = df1[df1['price']<50000]
print(price_under)
print(df2)
result6 = pd.merge(price_under,df2)
print(result6)
```
