# 보스턴 주택가격 딥러닝

```python
#보스턴 주택가격 데이터셋
from tensorflow.keras.datasets.boston_housing import load_data
(x_train,y_train),(x_test,y_test)=load_data(path='boston_housin.npz',
                                           test_split=0.2, seed=777)
print(x_train.shape,y_train.shape)
print(x_test.shape,y_test.shape)
```

```python
#데이터 전처리: 데이터의 표준화, 각 변수간의 값을 조정
'''
MinMax normalization: X=(x-min)/(max-min) => x/max
Robust normalization: X=(x-중간값)/(3분위값 1분위값)
Standardization:X=x-mean/std
'''
#평균
import numpy as np
mean=np.mean(x_train,axis=0) #평균값
print(mean[:10]) 
std=np.std(x_train,axis=0) #표준편차값
print(std[:10])
#표준화값을 설정
x_train=(x_train - mean)/std
x_test=(x_test - mean)/std
```

```python
#훈련데이터셋과 검증데이터셋 분리
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val=train_test_split(x_train,y_train,test_size=0.33,
                                            random_state=777)
#모델구성
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model=Sequential()
model.add(Dense(64, activation='relu', input_shape=(13,)))
model.add(Dense(32, activation='relu'))
model.add(Dense(1))
#mse: 평균제곱오차. (실제값-예측값)**2
#mae: 평균절대오차. |실제값-예측값|
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
#학습하기
history=model.fit(x_train,y_train,epochs=300, validation_data=(x_val,y_val))
model.evaluate(x_test,y_test)
```

```python
#그래프로 테스트 데이터 예측값, 테스트데이터의 실제값을 그래프로 작성
results=model.predict(x_test)
xval=range(len(x_test))

import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,5))
ax1=fig.add_subplot(1,2,1)
ax1.plot(xval,results, color='blue', label='predict_value')
ax1.set_title('predict value')
ax1.set_xlabel('data')
ax1.set_ylabel('price')
ax2=fig.add_subplot(1,2,2)
ax2.plot(xval,y_test,color='orange',label='real_value')
ax2.set_title('real value')
ax2.set_xlabel('data')
ax2.set_ylabel('price')
plt.show()
```

```python
fig=plt.figure(figsize=(10,5))
ax1=fig.add_subplot(1,1,1)
ax1.plot(xval,results, color='blue',label='predict_value')
ax1.plot(xval,y_test, color='orange',label='real_value')
ax1.set_title('predict and real value')
ax1.set_xlabel('data')
ax1.set_ylabel('price')
ax1.legend()
plt.show()
```

# 교차검증
#학습데이터로 학습할경우 과대적합이 발생될 우려가 높다
#과대적합을 방지하기 위한 수단

```python
교차검증의 종류
- 홀드아웃검증: 데이터셋을 무작위로 선택. 학습, 검증을 반복
= k-fold(k폴드)검증: 데이터를 k개의 그룹으로 분리
    한개는 검증, 나머지는 학습데이터로 사용
데이터 갯수가 적은 경우에 교차검증을 사용
```

```python
from tensorflow.keras.datasets.boston_housing import load_data
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
import numpy as np
```

```python
#데이터 로드
(x_train, y_train),(x_test,y_test)=load_data(path='boston_housing.npz',test_split=0.2,
                                            seed=777)
#데이터 표준화
mean=np.mean(x_train,axis=0)
std=np.std(x_train,axis=0)
x_train=(x_train-mean)/std
x_test=(x_test-mean)/std
```

```python
#K-Fold 교차검증 설정
k=3
#데이터를 3개로 분리
kfold=KFold(n_splits=k)
def get_model():
    mode=Sequential()
    model.add(Dense(64, activation='relu', input_shape=(13,)))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam',loss='mse',metrics=['acc'])
    return model
```

```python
#mae:평균절대오차
mae_list=[]
#kfold.split(x_train): 학습데이터를 k(3)개로 분리함
#train_index: 학습데이터로 사용한 인덱스들
#val_index: 검증데이터로 사용한 인덱스들
for train_index, val_index in kfold.split(x_train):
    #x_train_fold: 교차검증의 학습데이터
    #x_val_fold: 교차검증의 검증데이터
    x_train_fold,x_val_fold=x_train[train_index],x_train[val_index]
    y_train_fold,y_val_fold=y_train[train_index],y_train[val_index]
    model=get_model()
    #학습하기
    model.fit(x_train_fold, y_train_fold, epochs=300,
             validation_data=(x_val_fold, y_val_fold))
    #평가하기(손실함수값, 평가기준값(mae값))
    _, test_mae=model.evaluate(x_test,y_test)
    mae_list.append(test_mae)
```

```python
model.evaluate(x_test,y_test)
#결과조회하기
mae_list
print('최종결과:',np.mean(mae_list))
```

```python
#산점도를 이용하여 실제 주택가격과 예측가격 시각화하기
results=model.predict(x_test)
results[:10]

plt.figure(figsize=(5,5))
plt.plot(y_test,results,'b')
plt.plot([min(y_test), max(y_test)],[min(results),max(results)], ls='--',c='.3')
plt.xlabel('y_test')
plt.ylabel('results')
plt.show()
```

```python
#보스턴 주택가격 데이터 로드 
from tensorflow.keras.datasets import boston_housing
(train_X,train_Y), (test_X,test_Y)=boston_housing.load_data()

#데이터 전처리(표준화)
mean=train_X.mean(axis=0)
std=train_X.std(axis=0)
train_X=(train_X-mean)/std
test_X=(test_X-mean)/std

#회귀모델
model=Sequential([Dense(units=52,activation='relu',input_shape=(13,)),
                 Dense(39,activation='relu'),
                 Dense(26,activation='relu'),
                 Dense(1)])
model.summary()
model.compile(optimizer='adam',loss='mse',metrics='mae')

#학습데이터의 손실함수값, 검증데이터의 손실함수값을 그래프로 출력
history=model.fit(train_X,train_Y,epochs=25, batch_size=32,validation_split=0.25)
```

```python
#그래프로 출력
import matplotlib.pyplot as plt
fig=plt.figure(figsize=(10,5))
ax1=fig.add_subplot(1,2,1)
ax1.plot(history.history['loss'], 'b-',label='loss')
ax1.plot(history.history['val_loss'], 'r-',label='val_loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('loss')
ax1.legend()
ax2=fig.add_subplot(1,2,2)
ax2.plot(history.history['mae'], 'b-',label='mae')
ax2.plot(history.history['val_mae'], 'r-',label='val_mae')
ax2.set_title('train and val mae')
ax2.set_xlabel('epochs')
ax2.set_ylabel('mae')
ax2.legend()
plt.xlabel('Epochs')
plt.show()
```
