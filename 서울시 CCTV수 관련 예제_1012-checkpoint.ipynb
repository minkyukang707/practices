# CCTV수와 인구수를 이용한 데이터분석

```python
import pandas as pd 
```

```python
CCTV_Seoul=pd.read_csv('C:/pyy/01. CCTV_in_Seoul.csv',encoding='utf-8')
CCTV_Seoul.head()
```

```python
#'기관명'컬럼의 이름을 '구별'로 바꾸기
CCTV_Seoul.rename(columns={'기관명':'구별'},inplace=True)
CCTV_Seoul.head()
```

```python
#첫번째 컬럼의 이름을 '구별'로 컬럼명 변경하기
CCTV_Seoul.rename(columns={CCTV_Seoul.columns[0]:'구별'},inplace=True)
CCTV_Seoul.head()
```

```python
#서울시 인구 현황 읽기
pop_Seoul=pd.read_excel('C:/pyy/01. population_in_seoul.xls')
pop_Seoul.head()
```

```python
#엑셀에서 필요한 컬럼만 조회 
#header 인덱스: 2
pop_Seoul=pd.read_excel('C:/pyy/01. population_in_seoul.xls',header=2, 
                       usecols='B,D,G,J,N')
pop_Seoul.head()
```

```python
# 컬렴명을 구별, 인구수, 한국인, 외국인, 고령자 컬렴으로 변경하기 
pop_Seoul.rename(columns={pop_Seoul.columns[0]:'구별',
                          pop_Seoul.columns[1]:'인구수',
                          pop_Seoul.columns[2]:'한국인',
                          pop_Seoul.columns[3]:'외국인',
                          pop_Seoul.columns[4]:'고령자'},inplace=True)
pop_Seoul.head()
```

```python
#첫번째 행을 제거하기
pop_Seoul.drop([0],inplace=True)
pop_Seoul.head()
```

```python
#CCTV_Seoul 데이터에 최근증가율 컬럼추가 
#최근증가율: (2014년~2016년 CCTV수) / 2013년도 이전 CCTV수 * 100
#최근증가율이 높은 5개의구를 조회 
CCTV_Seoul['최근증가율']=(CCTV_Seoul['2014년']+CCTV_Seoul['2015년']+CCTV_Seoul['2016년'])/CCTV_Seoul['2013년도 이전']*100
CCTV_Seoul.sort_values(by='최근증가율',ascending=False).head()
```

```python
#외국인비율, 고령자비율 컬럼 추가하기 
pop_Seoul['외국인비율']=pop_Seoul['외국인']/pop_Seoul['인구수']*100
pop_Seoul['고령자비율']=pop_Seoul['고령자']/pop_Seoul['인구수']*100
pop_Seoul.head()
```

```python
#외국인비율이 높은 5개의 구 정보 출력
pop_Seoul.sort_values(by='외국인비율',ascending=False).head()
```

# CCTV 데이터와 인구데이터 합치고 분석

```python
data_result = pd.merge(CCTV_Seoul, pop_Seoul, on='구별')
data_result.head()
```

```python
#2013년도 이전, 2014~2016년 컬럼 제거하기
del data_result['2013년도 이전']
del data_result['2014년']
del data_result['2015년']
del data_result['2016년']
data_result.head()
```

```python
#인덱스를 구별컬럼으로 변경하기. 
data_result.set_index('구별',inplace=True)
data_result.head()
```

```python
#상관계수
data_result.corr(method='pearson')
```

```python
#고령자비율과 소계 두개의 피처간의 상관계수 구하기 
import numpy as np
np.corrcoef(data_result['고령자비율'], data_result['소계'])
```

```python
#외국인비율과 소계 간의 상관계수
np.corrcoef(data_result['외국인비율'], data_result['소계'])
```

```python
#인구수와 소계 간의 상관계수
np.corrcoef(data_result['인구수'], data_result['소계'])
```

```python
# matplot을 이용한 그래프 그리기
#CCTV갯수를 막대그래프로 작성하기 
import matplotlib.pyplot as plt
from matplotlib import font_manager,rc
plt.figure()
plt.rc('font',family='Malgun Gothic')
```

```python
data_result['소계'].plot(kind='barh',grid=True, figsize=(10,10))
plt.show()
```

```python
#그래프를 소계의 내림차순으로 수평막대 그래프 작성하기
data_result['소계'].sort_values().plot(kind='barh',grid=True, figsize=(10,10))
plt.show()
```

```python
# 인구수대비 CCTV비율 컬럼추가하기
# 소계/인구수 *100
# CCTV비율 순으로 수평막대 그래프 작성하기 

data_result['CCTV비율']=data_result['소계']/data_result['인구수']*100
data_result.head()
```

```python
data_result['CCTV비율'].sort_values().plot(kind='barh',grid=True, figsize=(10,10))
plt.show()
```

```python
# 산점도 그리기
# 인구수와 소계 산점도 작성하기
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.xlabel('인구수')
plt.ylabel('CCTV갯수')
plt.grid
plt.show()
```

```python
# 인구수와 소계 산점도, 회귀선 작성하기 
# fp1: 최소제곱법을 이용한 상수값
# 1: 직선 
fp1=np.polyfit(data_result['인구수'], data_result['소계'],1)
fp1
```

```python
f1=np.poly1d(fp1)
f1 # 인구수에 맞는 y값을 제공하는 함수
```

```python
fx=np.linspace(100000,700000,100)
plt.figure(figsize=(10,10))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx,f1(fx),ls='dashed',lw=3,color='g')
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid
plt.show()
```

```python
'''
polyfit: 각 차원의 최소거리에 맞는 기울기와 절편값을 리턴
직선: 1차원 : y=ax+b, b절편
곡선: 2차원 : y=ax**2+bx+c, c절편
곡선: 3차원 : y=ax**3+bx**2+cx+d, d절편
'''
```

```python
# 산점도+회귀선 , 산점도위에 점의 색상을 회귀선과의 거리로 표시하기 
# 회귀선을 위한 상수 
fp1=np.polyfit(data_result['인구수'], data_result['소계'],1)
# fp1상수값을 이용하여  y값을 계산하기 위한 함수
f1=np.poly1d(fp1)
fx=np.linspace(100000,700000,100) # x축값. 10만~70만까지 100등분
# f1(data_result['인구수']): 인구수에 맞는 회귀선 y값
# np.abs: 절대값 함수
data_result['오차']=np.abs(data_result['소계']-f1(data_result['인구수']))
data_result.head()
```

```python
df_sort=data_result.sort_values(by='오차',ascending=False)
df_sort.head()
```

```python
# c=data_result['오차']: 점의 색상을 오차에 맞도록 설정
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'],data_result['소계'],
           c=data_result['오차'], s=50)
plt.plot(fx,f1(fx),ls='dashed',lw=3,color='g')
```

```python
# c=data_result['오차']: 점의 색상을 오차에 맞도록 설정
plt.figure(figsize=(14,10))
plt.scatter(data_result['인구수'],data_result['소계'],
           c=data_result['오차'], s=50)
plt.plot(fx,f1(fx),ls='dashed',lw=3,color='g')

# 점에 해당하는 구의 이름, 오차가 높은 구 10개의 정보를 표시하기
for n in range(10):
    plt.text(df_sort['인구수'][n]*1.02, df_sort['소계'][n]*0.98,
            df_sort.index[n], fontsize=15)
plt.xlabel('인구수')
plt.ylabel('CCTY갯수')
plt.colorbar()
plt.grid()
plt.show()
```
