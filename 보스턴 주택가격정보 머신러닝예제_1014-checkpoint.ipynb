# 주택가격 머신러닝

<컬럼설명>
Town: 도시명
LON: 경도
LAT: 위도
CMEDV: 주택가격
CRIM: 1인당 범죄율
ZN: 25000평방피트를 초과하는 거주지역의 비율
INDUS: 비상업지역 토지비율
CHAS: 찰스강 경계
NOX: 10ppm당 일상화질소
RM: 주택 1가구당 평균 방의 개수
AGE: 1940년 이전의 건축된 주택의 비율 
DIS: 5개의 보스턴직업센터까지의 접근성 지수
RAD: 방사형 도로까지의 접근성 지수
TAX: 10000달러당 재산세율
PTRATIO: 타운별 학생/교사 비율
B: 흑인의 비율 
LSTAT: 모집단의 하위계층의 비율(%)

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing = pd.read_csv('C:/pyy/BostonHousing2.csv')
housing.head()
```

```python
#컬렴명CMEDV를 y로 변경
housing=housing.rename(columns={'CMEDV':'y'})
```

```python
#회귀 모델 학습을 위한 데이터셋 분리
#y, RM,LSTAT,NOX 컬럼의 산점도 작성하기
cols=['y','RM','LSTAT','NOX']
sns.pairplot(housing[cols])
plt.show()
#가격,y -> 양의 상관관계
#LSTAT, NOX -> 음의 상관관계
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
#독립속성선택
X=housing[['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO',
          'B','LSTAT']]
#종속속성
y=housing['y']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2, random_state=33)
```

```python
#회귀모델 생성
lr=linear_model.LinearRegression()
#학습하기
model=lr.fit(X_train,y_train)
#R2 score 출력
print(model.score(X_train,y_train))
print(model.score(X_test,y_test))
```

```python
#피처계수
print(lr.coef_)
```

```python
#RMSE score출력하기
from sklearn.metrics import mean_squared_error
from math import sqrt
y_pre=lr.predict(X_train)
rmse=sqrt(mean_squared_error(y_train,y_pre))
print(rmse)
```

```python
y_pre=lr.predict(X_test)
rmse=sqrt(mean_squared_error(y_test,y_pre))
print(rmse)
```
