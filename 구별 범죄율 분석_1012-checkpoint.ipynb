# 구별 범죄율 분석예제

```python
import pandas as pd 
import numpy as np
```

```python
crime_Seoul=pd.read_csv('C:/pyy/02. crime_in_Seoul.csv',thousands=',',encoding='euc-kr')
crime_Seoul.head()
```

```python
crime_Seoul.info()
```

```python
'''
공공데이터(https://www.data.go.kr/index.do)
에서 받은 경찰철_경찰관서 위치,주소 csv파일을 읽어 
경찰서의 위치 정보를 저장한다
경찰청_경찰관서 위치, 주소_20200409.csv
'''
police_state=pd.read_csv('C:/pyy/경찰청_경찰관서 위치, 주소_20200409 - 복사본.csv',
                        thousands=',',encoding='euc-kr')
police_state.head()
```

```python
# 서울 중부경찰서 정보 조회하기
police_state[police_state['지구대파출소']=='서울중부경찰서']
```

```python
# 서울지역 경찰서 정보만 선택하여 저장하기 
police_seoul=police_state[police_state['청']=='서울청']
police_seoul.head()
```

```python
# 경찰서의 이름을 서울xxx경찰서 형태
# 중부서 -> 서울중부경찰서
# 범죄현황데이터의 경찰서 이름을 police_seoul데이터의 경찰서 이름으로 변경하여 리스트로저장
police_name=[]
for name in crime_Seoul['관서명']:
    police_name.append('서울'+str(name[:-1])+'경찰서')
police_name
```

```python
# 경찰서의 주소정보를 리스트로 저장하기 
# police_seoul 데이터 -> 주소정보
# police_name 이용하여 police_seoul 데이터에서 주소값만 
# police_address 리스트에 저장하기

police_address =[]

for name in police_name:
    select_police = police_seoul[police_seoul['지구대파출소']==name]
    print(select_police['주소'])
    police_address.append(select_police.loc[:,'주소'])

police_address
```

```python
# police_address 값을 이용하여 구이름을 저장하기
gu_name = []

for name in police_address:
    tmp=name.str.split()
    tmp=tmp.tolist()
    tmp_gu=tmp[0]
    gu_name.append(tmp_gu[0])

gu_name
```

```python
# crime_Seoul 데이터에 구별 컬럼을 추가하기
crime_Seoul.head()
crime_Seoul['구별']=gu_name
crime_Seoul.head()
```

```python
# 구별 범죄건수들의 합계구하기
crime_Seoul.groupby('구별').sum()

crime_sum=pd.pivot_table(crime_Seoul, index='구별',
                        aggfunc=np.sum)
crime_sum
```

```python
# 강간, 강도, 살인, 절도, 폭력
# 검거율 컬럼 추가하기
# 검거율 = 검거/발생 *100
# 각각 범죄에 대한 검거율 컬럼 추가하기
# 검거수 컬럼들은 제거하기

crime_sum['강간검거율']=crime_sum['강간 검거']/crime_sum['강간 발생']*100
crime_sum['강도검거율']=crime_sum['강도 검거']/crime_sum['강도 발생']*100
crime_sum['살인검거율']=crime_sum['살인 검거']/crime_sum['살인 발생']*100
crime_sum['절도검거율']=crime_sum['절도 검거']/crime_sum['절도 발생']*100
crime_sum['폭력검거율']=crime_sum['폭력 검거']/crime_sum['폭력 발생']*100
crime_sum.head()
```

```python
del crime_sum['강간 검거']
del crime_sum['강도 검거']
del crime_sum['살인 검거']
del crime_sum['절도 검거']
del crime_sum['폭력 검거']
```

```python
crime_sum[['살인 발생','살인검거율']].head()
```

```python
#검거율 데이터 중 100보다 큰값은 100으로 변경하기 
col_list= ['강간검거율','강도검거율','살인검거율','절도검거율','폭력검거율']

for column in col_list:
    crime_sum.loc[crime_sum[column]>100, '강간검거율']=100

crime_sum.head()
```

```python
# **발생 컬럼명을 **으로 바꾸기
crime_sum.rename(columns={'강간 발생':'강간','강도 발생':'강도',
                         '살인 발생':'살인','절도 발생':'절도',
                         '폭력 발생':'폭력'},inplace=True)
crime_sum.head()
```

```python
# 범죄값들을 정규화 하기
# 기본적인 머신기능 모듈
from sklearn import preprocessing
```

```python
col=['강간','강도','살인','절도','폭력']
x=crime_sum[col].values
# 정규식
# MinMaxScaler() 0~1사이의 값으로 정규화 
min_max_scaler = preprocessing.MinMaxScaler()

x_scaled = min_max_scaler.fit_transform(x.astype(float))
crime_norm=pd.DataFrame(x_scaled, columns=col,
                       index=crime_sum.index)
crime_norm
```
