# 퍼셉트론 연습

```python
퍼셉트론: b+x1w1+x2w2=y
x1,x2: 입력값
w1,w2: 가중치
b: 편향
y=0:x1w1+x2w2 ,+b
1:x1w1+x2w2 >b 
```

```python
#AND게이트 구현하기 
import numpy as np

def AND(x1,x2): #1,0
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.8
    tmp=np.sum(w*x)+b #-0.5
    if tmp<=0:
        return 0
    else:
        return 1
```

```python
for xs in [(0,0),(1,0),(0,1),(1,1)]:
    y=AND(xs[0],xs[1])
    print(str(xs)+'=>'+str(y))
```

```python
#OR게이트 구현하기
def OR(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.2
    tmp=np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1
```

```python
for xs in [(0,0),(1,0),(0,1),(1,1)]:
    y=OR(xs[0],xs[1])
    print(str(xs)+'=>'+str(y))
```

```python
#NAND게이트 구현하기
def NAND(x1,x2):
    x=np.array([x1,x2])
    w=np.array([-0.5,-0.5])
    b=0.8
    tmp=np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1
```

```python
for xs in [(0,0),(1,0),(0,1),(1,1)]:
    y=NAND(xs[0],xs[1])
    print(str(xs)+'=>'+str(y))
```

```python
#XOR게이트 구현하기: 단일 퍼셉트론으로 구현불가, 다중퍼셉트론 필요
def XOR(x1,x2):
    s1=NAND(x1,x2)
    s2=OR(x1,x2)
    y=AND(s1,s2)
    return y
```

```python
for xs in [(0,0),(1,0),(0,1),(1,1)]:
    y=XOR(xs[0],xs[1])
    print(str(xs)+'=>'+str(y))
```

```python
#활성화 함수: 비선형 데이터로 변환하기 위한 함수
#계단함수: 0,1값을 리턴 
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.array(x>0, dtype=np.int)
```

```python
X=np.arange(-5.0,5.0,0.1)
print(np.array(X>0))
```

```python
Y=step_function(X)
plt.plot(X,Y)
plt.ylim(-0.1,1.1) #y축의 범위 지정
plt.show()
```

```python
#시그모이드: 0~1.0사이의 실수값
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
X=np.arange(-5.0,5.0,0.1)
Y=sigmoid(X)
plt.plot(X,Y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
#ReLU: 0~ 이상의 값
def relu(x):
    return np.maximum(0,x)
```

```python
x=np.arange(-5.0,5.0,0.1)
y=relu(x)
plt.plot(x,y)
plt.ylim(-1.0,5.5)
plt.show()
```

```python
#텐서플로우를 이용하여 OR게이트 구현하기 
import tensorflow as tf
import numpy as np
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.losses import mse
```

```python
tf.random.set_seed(777) #시작점을 랜덤하게 설정
data=np.array([[0,0],[1,0],[0,1],[1,1]]) #학습데이터
label=np.array([[0],[1],[1],[1]]) #정답
#퍼셉트론을 층으로 설정
model=Sequential()
#층추가
#activation: 활성화 함수 결정
model.add(Dense(1,input_shape=(2,), activation='linear'))
#Model 설정
#optimizer: GD(Gradient descent): 경사하강법 -> 모든데이터검토
#SGD(Stochastics gradient descent) => mini batch
#loss: 손실함수, 비용함수
#mse: 최소제곱함수
#metrics: 평가방법 설정
model.compile(optimizer=SGD(),loss=mse, metrics=['acc'])
#data: 학습데이터
#label: 정답
#epchs: 200번 실행
model.fit(data,label,epochs=500) #학습하기
model.get_weights() #가중치와 편향값을 리턴
```

```python
model.predict(data) #예측된값
model.evaluate(data,label) #평가, 손실함수값, 정확도
```
