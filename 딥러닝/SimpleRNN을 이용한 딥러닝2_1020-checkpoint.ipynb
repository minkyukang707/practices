# IMDB데이터셋을 SimpleRNN으로 분석

```python
from tensorflow.keras.datasets import imdb
num_words=10000
(X_train, y_train),(X_test,y_test) =imdb.load_data(num_words = num_words)
X_train.shape
```

```python
#패딩하기
from tensorflow.keras.preprocessing.sequence import pad_sequences
max_len=500
pad_X_train=pad_sequences(X_train, maxlen=max_len)
pad_X_test=pad_sequences(X_test, maxlen=max_len)
#분석하기
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense, Embedding
```

```python
model=Sequential()
model.add(Embedding(input_dim=num_words, output_dim=32))

'''
return_sequences: 기본값 False. 출력값을 반환여부. 순환값
dropout: 입력값을 drop 시키는 비율
recurrent_dropout : 순환값을 drop시키는 비율

'''
model.add(SimpleRNN(32,return_sequences =True, dropout=0.15, recurrent_dropout=0.15))
model.add(SimpleRNN(32))
model.add(Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy', metrics=['acc'])
model.summary()
```

```python
history=model.fit(pad_X_train, y_train, batch_size=32,epochs=15,validation_split=0.2)
```

```python
import matplotlib.pyplot as plt
his_dict=history.history
loss=his_dict['loss']
val_loss=his_dict['val_loss']
epochs=range(1,len(loss)+1)
fig=plt.figure(figsize=(10,5))
#훈련 및 검증손실 그리기
ax1=fig.add_subplot(1,2,1)
ax1.plot(epochs,loss,color='blue',label='train_loss')
ax1.plot(epochs,val_loss,color='orange',label='val_loss')
ax1.set_title('train and val loss')
ax1.set_xlabel('epochs')
ax1.set_ylabel('loss')
ax1.legend()
acc=his_dict['acc']
val_acc=his_dict['val_acc']
#훈련 및 검증손실 그리기
ax2=fig.add_subplot(1,2,2)
ax2.plot(epochs,acc,color='blue',label='train_acc')
ax2.plot(epochs,val_acc,color='orange',label='val_acc')
ax2.set_title('train and val acc')
ax2.set_xlabel('epochs')
ax2.set_ylabel('acc')
ax2.legend()
plt.show()
```

```python
SimpleRNN 결론
Embedding층만 사용한 모델과 정확도에 차이가 없다
SimpleRNN층을 사용하는 경우, 긴문장의 데이터 처리에 문제가 있다
=> 앞부분의 상태를 기억하지만 시점이 지날수록 기억이 없어진다
```

# Spam메일 정보분석

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
data=pd.read_csv('C:/pyy/spam.csv')
print('총 샘플의 수:',len(data))
data.head()
```

```python
data.info()
```

```python
#v1,v2컬럼을 제외한 나머지 컬럼을 제거하기
del data['Unnamed: 2']
del data['Unnamed: 3']
del data['Unnamed: 4']
```

```python
data.head()
```

```python
#v1 컬럼의 ham값은 0, spam값은 1로 변환
data['v1']=data['v1'].replace(['ham','spam'],[0,1])
data.head()
```

```python
#중복된 내용이 있는 경우 레코드를 한개만 남김. 중복제거하기
data.drop_duplicates(subset=['v2'], inplace=True)
data.info()
```

```python
#v1 데이터 값의 갯수 출력하기
data['v1'].value_counts()
data['v1'].value_counts().plot(kind='bar')
#학습데이터(메일내용), 정답데이터(spam여부)로 분리
X_data=data['v2']
y_data=data['v1']
print(len(X_data), len(y_data))
#분석을 위한 전처리
#토큰화하기
tokenizer=Tokenizer(oov_token='<OOV>')
tokenizer.fit_on_texts(X_data)
sequences=tokenizer.texts_to_sequences(X_data)
print(sequences[0])
word_to_index=tokenizer.word_index
print(word_to_index)
```

```python
total_cnt=len(word_to_index) #단어의 수
rare_cnt=0 #기준이하의 빈도수를 가진 단어의 갯수
total_freq=0 #훈련데이터 전체 단어의 빈도수
rare_freq=0 #기준이하의 빈도수를 가진 단어의 빈도수의 총합
threshold=2 #기준값
#단어: 빈도수 5개 조회
list(tokenizer.word_counts.items())[:5]
```

```python
for key, value in tokenizer.word_counts.items():
    total_freq=total_freq + value
    if(value < threshold): #단어의 등장 빈도수가 threshold보다 작으면 
        rare_cnt = rare_cnt +1
        rare_freq=rare_freq+1
print(total_freq)  
print('등장 빈도가 %s번 이하인 희귀 단어의 수:%s'%(threshold - 1, rare_cnt))
print('단어 집합(vocabulary)에서 희귀 단어의 비율:',(rare_cnt/total_cnt)*100)
print('전체 등장 빈도에서 희귀 단어 등장 빈도 비율:',(rare_freq/total_freq)*100)
```

```python
#메일의 길이 정보 조회하기
X_data=sequences
print('메일의 최대길이:%d'%max(len(l) for l in X_data))
print('메일의 평균길이:%f'%(sum(map(len,X_data))/len(X_data)))
max_len=189
data=pad_sequences(X_data, maxlen=max_len)
print('훈련 데이터의 크기(shape):',data.shape)
#학습데이터와 테스트데이터 분리하기
n_of_train=int(len(sequences)*0.8)
n_of_test=int(len(sequences)-n_of_train)
print(n_of_train,n_of_test)
```

```python
X_test=data[n_of_train:] #테스트데이터
y_test=np.array(y_data[n_of_train:]) #테스트데이터 레이블
X_train=data[:n_of_train] #학습데이터
y_train=np.array(y_data[:n_of_train]) #학습데이터 레이블
print(X_test.shape)
print(X_train.shape)
print(X_train[0])
```

```python
#SimpleRNN으로 스팸메일 분류하기
from tensorflow.keras.layers import SimpleRNN, Embedding, Dense
from tensorflow.keras.models import Sequential
model=Sequential()
model.add(Embedding(len(word_to_index)+1,32)) #임베딩 벡터의 차원은 32
model.add(SimpleRNN(32)) #RNN 셀의 hidden_size는 32
model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
```

```python
history=model.fit(X_train,y_train, epochs=10, batch_size=64, validation_split=0.2)
```

```python
print('테스트정확도:',model.evaluate(X_test,y_test))
```

```python
#그래프 출력
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'],'b-',label='loss')
plt.plot(history.history['val_loss'],'r--',label='val_loss')
plt.xlabel('Epoch')
plt.legend()
plt.subplot(1,2,2)
plt.plot(history.history['acc'],'g-',label='accuracy')
plt.plot(history.history['val_acc'],'k--',label='val_accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```
