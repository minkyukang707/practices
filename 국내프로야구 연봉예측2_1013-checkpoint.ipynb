# 2018년도 연봉예측 심화과정

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn import preprocessing
```

```python
picher_file_path='C:/pyy/picher_stats_2017.csv'
batter_file_path='C:/pyy/batter_stats_2017.csv'
picher=pd.read_csv(picher_file_path)
batter=pd.read_csv(batter_file_path)
```

```python
#1. picher데이터셋의 연봉(2018) 컬럼명을 y로 변경하기
picher=picher.rename(columns={'연봉(2018)':'y'})

#2.틴명을 one-hot인코딩하여 picher 데이터셋에 컬럼추가
team_encoding = pd.get_dummies(picher['팀명'])
picher=pd.concat([picher,team_encoding],axis=1)

#3. 팀명 컬럼삭제
picher=picher.drop('팀명',axis=1)

#4. 선수명,y컬럼 제외한 모든 컬럼을 X 독립변수에 저장 
X=picher[picher.columns.difference(['선수명','y'])]
X.head()
```

```python
#5. y컬럼을 종속변수 y로 저장
y=picher['y']
y.head()
```

```python
#6. X 정규화
# X=preprocessing.StandardSCaler().fix(X).transform(X)
pd.options.mode.chained_assignment=None # 과학적 표기방법 사용안함
```

```python
# 정규화 함수 
def standard_scaling(df,scale_columns):
    for col in scale_columns:
        s_mean=df[col].mean()
        s_std=df[col].std()
        df[col]=df[col].apply(lambda x:(x-s_mean)/s_std)
    return df
```

```python
scale_columns=['승','패','세','홀드','블론','경기','선발','이닝',
                             '삼진/9','볼넷/9','홈런/9','BABIP','LOB%','ERA',
                             'RA9-WAR','FIP','kFIP','WAR',
                             '연봉(2017)']
picher_df=standard_scaling(picher,scale_columns)
picher_df.head()
```

```python
#7. 훈련데이터/검증데이터 분리 
X=picher_df[picher_df.columns.difference(['선수명','y'])]
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2, random_state=19)
```

```python
#statsmodels 라이브러리의 OLS클래스로도 회귀분석 실행 가능
import statsmodels.api as sm
```

```python
#statsmodel 라이브러리로 회귀분석을 수행합니다
X_train=sm.add_constant(X_train)
model=sm.OLS(y_train, X_train).fit() # 회귀분석
model.summary()
```

R-Squared(결정계수) : 0~1 사이의 값 
독립변수의 변동량으로 설명되는 종속변수의 변동량 
R2=0.45인 경우, 종속변수의 변동은 독립변수 변동의 45%로 설명된다. 
->상관계수의 제곱과 같다
상관계수: 독립변수와 종속변수간의 선형관계를 수치적으로 표시, -1~1사이의 값 

adj R-Squared(수정 결정계수): 독립변수가 많아 지는 경우 결정계수의 같이 커질수있으므로
표본의 크기와 독립변수의 수를 고려하여 다시 계산된 결정계수

P>|t|: 각 피처의 검정통계량(t-statistics)이 유의미한지를 나타내는 p-value값
0.05미만인 피처가 회귀분석에 유의미한 피처들임
이분석에서는 WAR, 연봉(2017), 한화 3개의 피처의 p-value가 0.05미만
-> 회귀분석에서 유의미한 피처들

다중공선성: 변수간의 상관관계가 높아서 분석에 부정적인 영향을 미치는 것 
VIF('분상팽창요인계수: variance inflation factor')로 평가함 
일반적으로 10~15정도를 넘으면 다중공선성에 문제가있다고 판단

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif = pd.DataFrame()
vif['VIF Factor']=[variance_inflation_factor(X.values,i)\
                  for i in range(X.shape[1])]
vif['features']=X.columns
vif.round(1)
```

1. VIF계수가  높은 피처는 일단 제거. 유사한 피처 중 한개만 제거
2. 다시 모델을 실행해서 공선성검증

```python
#피처 재선정
X=picher_df[['FIP','WAR','볼넷/9','삼진/9','연봉(2017)']]
y=picher_df['y']
#학습데이터, 훈련데이터 분리
X_train,X_test, y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=19)
```

```python
#R2 Score출력하기, 과적합이 해소 
print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
```

```python
#RMSE 출력하기
#MSE score 제곱근
#MSE: 평균제곱오차
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
y_predictions=lr.predict(X_train)
print(sqrt(mean_squared_error(y_train, y_predictions)))
y_predictions=lr.predict(X_test)
print(sqrt(mean_squared_error(y_test, y_predictions)))
```
