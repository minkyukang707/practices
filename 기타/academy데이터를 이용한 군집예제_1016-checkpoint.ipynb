# KMeans 예제

```python
import pandas as pd 
data=pd.read_csv('C:/pyy/academy1.csv')
data.info()
```

```python
#KMeans 알고리즘을 이용한 모델 생성하기
from sklearn.cluster import KMeans
model=KMeans(n_clusters=3)
#학습하기
model.fit(data.iloc[:,1:])
#예측하기
aaa=model.fit_predict(data.iloc[:,1:])
print('클러스터 레이블:',model.labels_)
```

```python
import mglearn
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font',family='Malgun Gothic') #현재 폰트 변경 설정
mglearn.plots.plot_kmeans_algorithm()
plt.show()
```

```python
#경계선 그리기
mglearn.plots.plot_kmeans_boundaries()
plt.show()
```

```python
mglearn.discrete_scatter(data.iloc[:,1],data.iloc[:,2],model.labels_)
plt.legend(['클러스터 0','클러스터 1','클러스터 2'],loc='best')
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.show()
```

```python
#Q: 국어점수100, 영어점수 80인 새로운 학생이 입학
#이 학생은 몇번 클러스테 포함되어야 하는지?
model.predict([[100,80]])
#국어점수 60, 영어점수50인 새로운학생이 입학
model.predict([[60,50]])
```

```python
#academy2 파일을 읽어서 data에 저장
#KMeans모델을 생성하여 학습시키고 3개의 클러스터로 분류
#국어점수 100, 영어점수80
#수학점수 70, 과학점수 70, 학업성취도 70
#클러스터 번호 출력하기
data=pd.read_csv('C:/pyy/academy2.csv')
model=KMeans(n_clusters=3)
data.head()
```

```python
#학습하기
model.fit(data.iloc[:,1:])
print(model.labels_)
print('\n')
print(data.iloc[:,1:])
```

```python
#예측하기
aaa=model.fit_predict(data.iloc[:,1:])
print('클러스터 레이블:',model.labels_)
print('\n')
model.predict([[100,80,70,70,70]])

#학생들의 클러스터 정보 출력하기
for no, cla in enumerate(model.labels_):
    print(data.iloc[no].tolist(),cla)
```

# DBSCAN 모델적용

```python
import pandas as pd 
data=pd.read_csv('C:/pyy/academy2.csv')
```

```python
#DBSCAN 알고리즘을 이용한 모델 생성하기
from sklearn.cluster import DBSCAN
model=DBSCAN()
#학습하기
model.fit(data.iloc[:,1:])
clusters=model.fit_predict(data.iloc[:,1:])
print('클러스터:',clusters)

mglearn.discrete_scatter(data.iloc[:,1],data.iloc[:,2],model.labels_)
plt.legend(['클러스터 0','클러스터 1','클러스터 2'], loc='best')
plt.xlabel('국어점수')
plt.ylabel('영어점수')
plt.show()
```

```python
KMeans 알고리즘
1.성능이 좋다. 군집에 많이 사용됨
2. 데이터의 클러스터가 원의 형태로 가정하여 클러스터를 구분
3. 클러스터의 개수를 수동으로 설정

DBSCAN 알고리즘
1.밀집도를 기준으로 클러스터 구분
```
