# auto-mpg데이터를 이용한 머신러닝예제

```python
# 기본라이브러리 불러오기
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df=pd.read_csv('C:/pyy/auto-mpg.csv',header=None)
```

```python
# 열이름 지정 
df.columns=['mpg','cylinders','displacement','horsepower','weight','acceleration',
           'model year','origin','name']
df.head()
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.horsepower.unique())
```

```python
# horsepower 컬럼의 ?값을 결측값(np.nan)으로 변경하기
df['horsepower'].replace('?',np.nan,inplace=True)
df['horsepower'].isnull().sum()
```

```python
print(df.info())
```

```python
# 누락데이터행을 삭제 
df.dropna(subset=['horsepower'],axis=0, inplace=True)
print(df.info())
```

```python
# 10까지의 컬럼을 조회하도록 설정
pd.set_option('display.max_columns',10)
print(df.describe()) #horsepower의 컬럼이 나오지않음. 숫자형이 아니므로 
```

```python
# horsepower의 데이터 성질을 실수형으로 변환
df['horsepower']=df['horsepower'].astype('float')
print(df.describe())
```

```python
# 분석에 활용할 속성(열)을 선택
# 연비,실린더, 출력, 중량 선택
ndf=df[['mpg','cylinders','horsepower','weight']]
ndf.info()
```

```python
# matplot으로 산점도 그리기
# mpg,weight 변수의 산점도 작성하기 
ndf.plot(kind='scatter', x='weight',y='mpg',c='coral',s=10,figsize=(10,5))
plt.show()
```

```python
# seaborn으로 산점도 그리기
fig=plt.figure(figsize=(10,5))
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)
sns.regplot(x='weight',y='mpg',data=ndf,ax=ax1) # 회귀선 표시
sns.regplot(x='weight',y='mpg',data=ndf,ax=ax2, fit_reg=False) # 회귀선 미표시
plt.show()
```

```python
# jointplot 그리기
sns.jointplot(x='weight',y='mpg',data=ndf) # 회귀선없음
sns.jointplot(x='weight',y='mpg', kind='reg',data=ndf) # 회귀선표시
plt.show()
```

```python
# seaborn pairplot 
sns.pairplot(ndf,kind='reg')
plt.show()
```

```python
# 데이터셋 구분 - 훈련용(train), 검증용(test)
# 독립변수 : X
X=ndf[['weight']]
print(type(X))
#종속변수 Y
Y=ndf['mpg']
print(type(Y))
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 훈련용데이터, 검증용데이터 분리(7:3)
# random_state=10: 추출사용되는 상수. 시드값
# 분리되는 데이터의 유지성을 위한 설정 
X_train, X_test, y_train, y_test = train_test_split\
(X,Y,test_size=0.3, random_state=10)
```

```python
print('train data 개수:', len(X_train))
print('test data 개수:', len(X_test))
X_train[:10]
```

```python
# 단순회귀분석 모형 - sklearn 사용
# sklearn 라이브러리에서 선형회귀분석 모듈 가져오기 
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
X=ndf[['cylinders','horsepower','weight']]
Y=ndf['mpg']
```

```python
# train data와 test data로 구분
X_train, X_test, y_train, y_test = train_test_split\
(X,Y,test_size=0.3, random_state=10)
```

```python
print('훈련 데이터:',X_train.shape)
print('검증 데이터:',X_test.shape)
```

```python
# X_train 훈련시키기
lr.fit(X_train, y_train)
```

```python
# 결정계수(R-제곱) 조회하기
r_square = lr.score(X_test, y_test)
print('결정계수:',r_square)
```

```python
# 회귀식의 기울기
print('회귀식의 기울기(a):', lr.coef_)
# 회귀식의 y절편
print('회귀식의 y절편(b):',lr.intercept_)
```

```python
# test데이터로 예측하기
y_hat=lr.predict(X_test)

#예측된 데이터와 실데이터를 kdeplot으로 출력하기
plt.figure(figsize=(10,5))
ax1=sns.kdeplot(y_test, label='y_test')
ax2=sns.kdeplot(y_hat, label='y_hat',ax=ax1)
plt.legend()
plt.show()
```

```python
# 원 데이터로 예측하기
y_har=lr.predict(X)

# 예측된 데이터와 실데이터를 kdeplot으로 출력하기
plt.figure(figsize=(10,5))
ax1=sns.kdeplot(Y, label='Y')
ax2=sns.kdeplot(y_hat, label='y_hat',ax=ax1)
plt.legend()
plt.show()
```
