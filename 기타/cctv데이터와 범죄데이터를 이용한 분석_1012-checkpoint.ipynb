```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
plt.rc('font', family ='Malgun Gothic')
```

```python
# CCTV데이터 
CCTV_Seoul = pd.read_csv('C:/pyy/01. CCTV_in_Seoul.csv',  encoding='utf-8')
CCTV_Seoul.rename(columns={"기관명" : '구별'}, inplace=True)
del CCTV_Seoul["2013년도 이전"]
del CCTV_Seoul["2014년"]
del CCTV_Seoul["2015년"]
del CCTV_Seoul["2016년"]
```

```python
CCTV_Seoul.head()
```

```python
# 범죄데이터 
crime_Seoul = pd.read_csv('C:/pyy/02. crime_in_Seoul.csv',thousands=',', encoding='euc-kr')

police_state = pd.read_csv('C:/pyy/경찰청_경찰관서 위치, 주소_20200409 - 복사본.csv', thousands=',', encoding='euc-kr')
police_seoul = police_state[police_state["청"]=='서울청']
police_name = []
for name in crime_Seoul["관서명"] :
   police_name.append('서울' + str(name[:-1]) + '경찰서')
police_address = []

for name in police_name :
    select_police = police_seoul[police_seoul['지구대파출소']==name]
    police_address.append(select_police.loc[:,'주소'])

gu_name=[]
for name in police_address:
    tmp = name.str.split()
    tmp = tmp.tolist()
    tmp_gu = tmp[0]
    gu_name.append(tmp_gu[0])

crime_Seoul["구별"]=gu_name   
crime_Seoul.head()
```

```python
# 범죄검거율 컬럼생성
crime_Seoul["발생"] = \
crime_Seoul["살인 발생"] + crime_Seoul["강도 발생"] + crime_Seoul["강간 발생"] +crime_Seoul["절도 발생"] +crime_Seoul["폭력 발생"]
crime_Seoul["검거"] = \
crime_Seoul["살인 검거"] + crime_Seoul["강도 검거"] + crime_Seoul["강간 검거"] +crime_Seoul["절도 검거"] +crime_Seoul["폭력 검거"]
crime_Seoul["검거율"] = crime_Seoul["검거"]/crime_Seoul["발생"] * 100
```

```python
# CCTV데이터와 범죄데이터를 구를 기준으로 병합
data_result = pd.merge(CCTV_Seoul, crime_Seoul, on='구별')
data_result.head()
```

```python
#경찰서별 전체 범죄 검거율 컬럼을 생성하고, 검거율별로 수평막대그래프로 작성하기 
data1 =  pd.merge(CCTV_Seoul, crime_Seoul, on='구별')
data1.info()
data1.set_index('관서명', inplace=True)
data1.sort_values(by="검거율")
plt.figure()
data1['검거율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
#구별 전체 범죄 검거율 컬럼을 생성하고, 검거율별로 수평막대그래프로 작성하기 
data2 = pd.pivot_table(data_result,index="구별",aggfunc=np.mean)
data2["검거율"]
plt.figure()
data2['검거율'].sort_values().plot(kind='barh', grid=True, figsize=(10,10))
plt.show()
```

```python
#경찰서별 범죄발생건수, CCTV 갯수를 산점도와 회귀선으로 출력하기.
data3 =  pd.merge(CCTV_Seoul, crime_Seoul, on='구별')
data3.set_index('관서명', inplace=True)
fp1 = np.polyfit(data3['소계'], data3['발생'], 1)
data3['소계']
f1 = np.poly1d(fp1)
fx = np.linspace(500, 3000, 100)
data3['오차'] = np.abs(data3['발생'] - f1(data3['소계']))

df_sort = data3.sort_values(by='오차', ascending=False)
plt.figure(figsize=(14,10))
plt.scatter(df_sort['소계'],df_sort["발생"],c=df_sort['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
for n in range(10):
    plt.text(df_sort['소계'][n]*1.02, df_sort['발생'][n]*0.98, 
             df_sort.index[n], fontsize=15)
plt.xlabel('CCTV 갯수')
plt.ylabel('전체범죄 발생수')
plt.colorbar()
plt.grid()
plt.show()
```

```python
#경찰서별 검거율과, CCTV 갯수를 산점도와 회귀선으로 출력하기.
data4 =  pd.merge(CCTV_Seoul, crime_Seoul, on='구별')
data4.set_index('관서명', inplace=True)
fp1 = np.polyfit(data4['소계'], data4['검거율'], 1)
data4['소계']
f1 = np.poly1d(fp1)
fx = np.linspace(500, 3000, 100)
data4['오차'] = np.abs(data4['검거율'] - f1(data4['소계']))
df_sort = data4.sort_values(by='오차', ascending=False)
plt.figure(figsize=(14,10))
plt.scatter(df_sort['소계'],df_sort["검거율"],c=df_sort['오차'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
for n in range(10):
    plt.text(df_sort['소계'][n]*1.001, df_sort['검거율'][n]*0.999, 
             df_sort.index[n], fontsize=15)
plt.xlabel('CCTV 갯수')
plt.ylabel('전체범죄 검거율')
plt.colorbar()
plt.grid()
plt.show()
```
