```python
import tensorflow as tf
import matplotlib.pyplot as plt
#image_path: 이미지파일이 저장된 경로
#tf.keras.utils.get_file::http:/bit.ly/33U6mH9데이터를 cat.jpg
image_path=tf.keras.utils.get_file('cat.jpg','http://bit.ly/33U6mH9')
image_path
```

```python
#imread: 이미지값을 숫자 형태로 리턴
image=plt.imread(image_path)
print(image.shape)
print(image[0])
titles=['RGB Image','Red channel','Green channel','Blue channel']
```

```python
from numpy import array, zeros_like
def channel(image,color): #0
    if color not in (0,1,2): return image
    c=image[...,color]
    print(c)
    #zeros_like: c배열과 같은 형태의 배열이고 전부 0인 값을 가진배열 생성
    z=zeros_like(c)
    #transpose: 배열의 형태를 변경
    return array([(c,z,z),(z,c,z),(z,z,c)][color]).transpose(1,2,0)
```

```python
colors=range(-1,3) #-1,-2,~0,1,2
fig,axes=plt.subplots(1,4,figsize=(13,3))
objs=zip(axes,titles,colors)
for ax,title,color in objs:
    ax.imshow(channel(image,color))
    ax.set_title(title)
    ax.set_xticks(())
    ax.set_yticks(())
plt.show()
```
