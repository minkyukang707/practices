# 셀레니움예제

```python
from selenium import webdriver 
import time
import urllib.request as req
import os 
```

```python
driver = webdriver.Chrome('C:/pyy/chromedriver')
driver.get('https://search.daum.net/search?nil_suggest=sugsch&w=img&DA=GIQ&sq=%EC%9E%A5%EB%AF%B8&o=1&sugo=15&q=%EC%9E%A5%EB%AF%B8%22')
#장미검색
time.sleep(3)
images=driver.find_elements_by_css_selector("#imgList > div > a >img")
img_url = [] # 이미지 url 정보를 저장하고 있는 리스트
for image in images: 
    #image : 이미지 태그
    #get_attribute('src') : img 태그의 속성의 값 
    #url = https://search4.kakaocdn.net/argon/0x200_85_hr/IKDZidUywmD
    url = image.get_attribute('src')
    img_url.append(url)
print(img_url)
driver.quit()
img_folder = './img' #이미지를 저장할 폴더 선택 
#isdir : 해당 폴더가 존재하는지
#해당 폴더가 없으면 폴더를 새로 생성하는 조건문 
if not os.path.isdir(img_folder):
    os.mkdir(img_folder) #폴더를 생성하는 함수 
```

```python
# enumertae : 리스트의 값과 인덱스를 조회
# index: 리스트의 인덱스 
# link: 리스트의 값 

for index, link in enumerate(img_url):
    req.urlretrieve(link,f'./img/{index}.jpg')
```

```python
#네이버 홈화면을 이미지 파일로 저장하기 
url = 'http://www.naver.com/'
driver.get(url)
#현재 브라우저의 화면을 이미지 저장하기
#naverhome.png: 저장되는 이미지의 이름
driver.save_screenshot('naverhome.png')
driver.quit()
```
