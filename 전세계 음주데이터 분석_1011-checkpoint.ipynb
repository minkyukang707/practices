# 상관계수 시각화 

```python
#drinks 데이터셋의 기본정보 조회하기
import pandas as pd 
```

```python
drinks=pd.read_csv('C:/pyy/drinks.csv')
drinks.info()
drinks.head()
```

```python
#피처간의 상관관계
#맥주, 와인 상관관계 조회하기
#'beer_servings','wine_servings' 두 피처간의 상관계수를 계산
#pearson은 상관계수를 구하는 계산 방법 중 하나, 가장 널리 쓰이는 방법
corr=drinks[['beer_servings','wine_servings']].corr(method='pearson')
corr
```

```python
#'beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol'
#상관계수 출력하기
#조회되는 컬럼의 갯수는 설정안함 
pd.set_option('display.max_columns',None)
```

```python
cols=['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']
corr=drinks[cols].corr(method='pearson')
corr
```

```python
#상관계수 시각화 하기
#corr 행렬 히트맵을 시각화
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#그래프 출력을 위한 cols 이름을 축약
cols_view=['beer','spirit','wine','alcohol']
sns.set(font_scale=1.5)
hm=sns.heatmap(corr.values,cbar=True, annot=True, square=True, fmt='.2f',
              annot_kws={'size':15},
              yticklabels=cols_view,
              xticklabels=cols_view)
plt.show()
```

```python
#시각화 라이브러리를 이용한 피처간의 scatter plot을 출력 
sns.pairplot(drinks[['beer_servings','spirit_servings','wine_servings','total_litres_of_pure_alcohol']],height=2.5)
plt.show()
```

```python
#각 피처의 결측값의 갯수 출력
drinks.isnull().sum()
```

```python
#대륙의 결측값을 OT로 채우기
drinks['continent']=drinks['continent'].fillna('OT')
```

```python
#대륙별 국가수를 출력하기
print(drinks['continent'].value_counts()) #방법1
```

```python
print(drinks.groupby('continent').count()['country']) #방법2
```

```python
#대륙별 국가의 갯수를 파이그래프로 출력하기 
#plt.pie 함수
import matplotlib.pyplot as plt

labels = drinks['continent'].value_counts().index.tolist()
explode=(0,0,0,0.2,0,0)
plt.pie(drinks['continent'].value_counts(), labels=labels,autopct='%.0f%%',
       explode=explode,shadow=True)
plt.title("null data to OT")
plt.show()
```

```python
#대륙별 spirit_servings의 평균, 최소, 최대, 합계를 출력
result=drinks.groupby('continent').spirit_servings.agg(['mean','min','max','sum'])
result #한번에 모든 함수적용
```

```python
#total_litres_of_pure_alcohol: 알콜량
#알콜량이 전체 평균보다 많은 대륙을 출력하기
#전체평균
total_mean = drinks.total_litres_of_pure_alcohol.mean()
total_mean
```

```python
#대륙별 평균
continent_mean=drinks.groupby('continent')['total_litres_of_pure_alcohol'].mean()
continent_mean
```

```python
#대륙별평균값 >= 전페평균 조건에 맞는 데이터 출력
continent_over_mean=continent_mean[continent_mean>=total_mean]
print(continent_over_mean)
```

```python
#평균 beer_servings이 가장 높은 대륙을 출력 
#대륙별 평균 beer_servings
#Series.idxmax() : 최대값을 가지는 index값 리턴 
cont_beer_mean = drinks.groupby('continent').beer_servings.mean()
print(cont_beer_mean)
print(type(cont_beer_mean))
```

```python
drinks.groupby('continent').beer_servings.mean().idxmax() #가장 높은대륙
```

```python
#평균 beer_servings이 가장 낮은 대륙을 출력
drinks.groupby('continent').beer_servings.mean().idxmin()
```

```python
#대륙별 spirit_servings의 평균, 최소,최대,합계를 시각화
import numpy as np 
```

```python
n_groups = len(result.index)
result['mean'].tolist() # 대륙별 spirit_servings 평균값 리스트
```

```python
means = result['mean'].tolist()
mins = result['min'].tolist()
maxs = result['max'].tolist()
sums = result['sum'].tolist()
index=np.arange(n_groups)
```

```python
bar_width=0.1
rects1=plt.bar(index,means,bar_width,color='r',label='Mean')
rects2=plt.bar(index+bar_width,mins,bar_width,color='g',label='Min')
rects3=plt.bar(index+bar_width*2,maxs,bar_width,color='b',label='Max')
rects4=plt.bar(index+bar_width*3,sums,bar_width,color='y',label='sum')
plt.xticks(index,result.index.tolist())
plt.legend(loc='best')
plt.show()
```

```python
#대륙별 total_litres_of_pure_alcohol의 평균 시각화
#대륙별 알콜섭취량 평균
#전체평균 추가하기
continent_mean=drinks.groupby('continent')['total_litres_of_pure_alcohol'].mean()
continent_mean #대륙별 평균 
```

```python
total_mean=drinks.total_litres_of_pure_alcohol.mean() #전체평균
total_mean
```

```python
#대륙명
continents = continent_mean.index.tolist()
continents.append('Mean') #대륙명에게 전체평균에 해당하는 index추가 
x_pos = np.arange(len(continents))
alcohol = continent_mean.tolist()
alcohol.append(total_mean) #전체평균 데이터를 추가 
bar_list = plt.bar(x_pos, alcohol, align='center', alpha=0.5)
```

```python
#마지막 전체평균 막대의 색을 빨간색으로 변경
bar_list[len(continents)-1].set_color('r')
```

```python
#전체평균의 값에 선그래프로 표시하기
plt.plot([0.,6],[total_mean,total_mean],'k--')
#bar_list[-1:2].set_color('r')
plt.xticks(x_pos, continents) #x축에 대륙코드설정
plt.ylabel('total_litres_of_pure_alcohol')
plt.title('total_litres_of_pure_alcohol by Continent')
plt.show()
```

```python
#변경사항 합치기
continents = continent_mean.index.tolist()
continents.append('Mean') #대륙명에게 전체평균에 해당하는 index추가 
x_pos = np.arange(len(continents))
alcohol = continent_mean.tolist()
alcohol.append(total_mean) #전체평균 데이터를 추가 

bar_list = plt.bar(x_pos, alcohol, align='center', alpha=0.5)
#마지막 전체평균 막대의 색을 빨간색으로 변경
bar_list[len(continents)-1].set_color('r')

#전체평균의 값에 선그래프로 표시하기
plt.plot([0.,6],[total_mean,total_mean],'k--')
#bar_list[-1:2].set_color('r')
plt.xticks(x_pos, continents) #x축에 대륙코드설정
plt.ylabel('total_litres_of_pure_alcohol')
plt.title('total_litres_of_pure_alcohol by Continent')
plt.show()
```

```python
# 대륙별 beer_servings의 합계를 막대그래프로 시각화 
# EU막대의 색상을 빨간색으로 변경하기
# 전체 맥주 소비량 합계의 평균을 구해서 막대그래프에 추가
# 평균선을 출력하기. 막대 색상은 노란색
# 평균선은 검정색('k--')
beer_sum = drinks.groupby('continent')['beer_servings'].sum()
beer_sum
```

```python
beer_mean=beer_sum.mean()
beer_mean #전체평균
```

```python
continents = beer_sum.index.tolist()
continents.append('Mean')
continents # 대륙명 + 평균
```

```python
x_pos = np.arange(len(continents))
alcohol=beer_sum.tolist()
alcohol.append(beer_mean)
alcohol # 대륙명 값들을 저장
```

```python
bar_list=plt.bar(x_pos, alcohol, align='center',alpha=0.5)
bar_list[2].set_color('r') # EU의 막대그래프의 색을 빨간색으로 변경
bar_list[continents.index('Mean')].set_color('y') # Mean의 막대그래프의 색을 노란색으로 변경
plt.xticks(x_pos, continents) # x축에 대륙코드 설정
plt.plot([0.,6], [beer_mean,beer_mean],'k--')
plt.ylabel('beer_servings')
plt.title('beer_servings by Continent')
plt.show()
```
