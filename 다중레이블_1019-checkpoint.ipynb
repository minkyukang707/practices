```python
다중레이블: 이미지분석의 결과가 2개이상인 경우
빨강, 파랑, 드레스, 신발 => 손실함수: binary_crossentropy
1 0 1 0 => 활성화함수 : 시그모이드
0 1 1 0
다중클래스-> 손실함수: categorical_crossentropy
사과 배 오렌지 -> 활성화함수 : 소프트맥스
0 1 0

소프트맥스는 가장 높은 확률의 클래스만 강조되므로 다중레이블에서는 부적합
```

```python
import pandas as pd 
DATA_PATH='C:/pyy'
```

```python
train_df=pd.read_csv(DATA_PATH+'/train.csv')
val_df=pd.read_csv(DATA_PATH+'/val.csv')
test_df=pd.read_csv(DATA_PATH+'/test.csv')
train_df.info()
```

```python
train_df['image'] 컬럼을 읽어서 분석함 
-> 5578개의 이미지를 메모리에 로드하기 어려우므로 이미지 제너레이터를 이용
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
#rescale: 이미지데이터/255 -> 정규화
train_datagen=ImageDataGenerator(rescale=1./255)
val_datagen=ImageDataGenerator(rescale=1./255)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Flatten,Conv2D,MaxPool2D,Dropout
```

```python
#기본적인 딥러닝으로 분석
model=Sequential()
model.add(Flatten(input_shape=(112,112,3))) #1차원 배열(112*1112*3)
model.add(Dense(128,activation='relu')) #128개의 출력
model.add(Dense(64,activation='relu')) #64개의 출력
model.add(Dense(11,activation='softmax')) #11개의 출력. 다중레이블

model.compile(optimizer='adam',loss='binary_crossentropy',
             metrics=['acc'])
batch_size=32
class_col=['black','blue','brown','green','red','white','dress','shirt','pants',
          'shorts','shoes']

'''
flow_from_dataframe: dataframe을 기준으로 이미지 읽기
flow_from_directory: directory를 기준으로 이미지 읽기
'''
train_df.info()
```

```python
train_generator=train_datagen.flow_from_dataframe(
dataframe=train_df,#사용할 데이터 프레임
directory=None, #이미지를 저장하고 있는 폴더. 여기는 필요없음. 이미지파일이 경로포함
x_col='image', #학습데이터를 저장하고 있는 컬럼. 이미지
y_col=class_col, #레이블데이터를 저장하는 컬럼
target_size=(112,112), #이미지크기
color_mode='rgb', #색상설정
class_mode='raw', #배열
batch_size=batch_size, #생성하는 이미지 개수
shuffle=True, #이미지를 섞어서 생성
seed=42) #랜덤시드

val_generator=val_datagen.flow_from_dataframe(
dataframe=val_df,#사용할 데이터 프레임
directory=None, #이미지를 저장하고 있는 폴더. 여기는 필요없음. 이미지파일이 경로포함
x_col='image', #학습데이터를 저장하고 있는 컬럼. 이미지
y_col=class_col, #레이블데이터를 저장하는 컬럼
target_size=(112,112), #이미지크기
color_mode='rgb', #색상설정
class_mode='raw', #배열
batch_size=batch_size, #생성하는 이미지 개수
shuffle=True) #이미지를 섞어서 생성
```

```python
import matplotlib.pyplot as plt
batch=next(train_generator)
image,level=batch[0],batch[1]
print(image[0])
print(level[1])
plt.imshow(image[0])
```

```python
#데이터갯수/32 나눈몫
def get_steps(num_samples,batch_size):
    if(num_samples % batch_size)>0:
        return(num_samples//batch_size)+1
    else:
        return num_samples//batch_size
```

```python
#steps_per_epoch: epoch시 이동하는 갯수
history=model.fit(train_generator,
                 steps_per_epoch=get_steps(len(train_df),batch_size),
                 validation_data=val_generator,
                 validation_steps=get_steps(len(val_df),batch_size),epochs=10)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'],'b-',label='loss')
plt.plot(history.history['val_loss'],'r--',label='val_loss')
plt.xlabel('Epoch')
plt.legend()
plt.subplot(1,2,2)
plt.plot(history.history['acc'],'g-',label='accuracy')
plt.plot(history.history['val_acc'],'k--',label='val_accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

# 테스트데이터를 이용하여 예측

```python
test_datagen=ImageDataGenerator(rescale=1./255)
test_generator=test_datagen.flow_from_dataframe(
dataframe=test_df,#사용할 데이터 프레임
directory=None, #이미지 폴더 정보없음
x_col='image', #테스트 데이터 사용할 이미지 정보를 저장한 컬럼명
y_col=None, #정답값을 설정하지 않음
target_size=(112,112), #이미지크기
color_mode='rgb', #색상설정
class_mode=None, #정답값 설정안함
batch_size=batch_size, #32개씩 
shuffle=False) #이미지를 섞지않음

preds=model.predict(test_generator,steps=32) #예측하기
off=0
do_preds=preds[off:off+8] #8개만 조회하기
do_preds
```

```python
for i,pred in enumerate(do_preds):
    plt.subplot(2,4,i+1) #2행 4열의 형태로 이미지 출력
    prob=zip(class_col,list(pred))
    #prob:('black',0.001..)
    #0.001.. 숫자값(예측한 확율)의 역순으로 정렬
    #큰값을 가진 데이터 2개만 저장
    prob=sorted(list(prob),key=lambda z:z[1],reverse=True)[:2]
    image=plt.imread(test_df['image'][i+off]) #이미지데이터 읽기
    plt.imshow(image) #이미지로 화면에 출력
    #prob[0][0]: black 중 하나 확률이 가장 높은 값의 종류
    #prob[0][1]: 확률값 중 가장 큰값
    #*100한 값의 소숫점이하 2자리까지만 반올림해서 출력
    #prob[1][0]: black중 하나 확률이 2번째로 높은값의 종류
    #prob[1][1]: 확률값 중 2번쨰로 큰 값
    #*100한 값의 소숫점이하 2자리까지만 반올림해서 출력
    plt.title(f'{prob[0][0]}:{round(prob[0][1]*100,2)}%\n{prob[1][0]}:{round(prob[1][1]*100,2)}%')
    plt.tight_layout()
plt.show()
    
```

# CNN을 이용하여 분석하기

```python
model=Sequential([
    Conv2D(input_shape=(112,112,3),kernel_size=(3,3),
          filters=32,padding='same',activation='relu'),
    Conv2D(kernel_size=(3,3),
          filters=64,padding='same',activation='relu'),
    MaxPool2D(pool_size=(2,2)),
    Dropout(rate=0.5),
    Conv2D(kernel_size=(3,3),filters=128,padding='same',activation='relu'),
    Conv2D(kernel_size=(3,3),filters=128,padding='valid',activation='relu'),
    MaxPool2D(pool_size=(2,2)),
    Dropout(rate=0.5),
    Flatten(),
    Dense(units=512,activation='relu'),
    Dropout(rate=0.5),
    Dense(units=256,activation='relu'),
    Dropout(rate=0.5),
    Dense(units=11,activation='sigmoid')
])

model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['acc'])
history=model.fit(train_generator,
                 steps_per_epoch=get_steps(len(train_df),batch_size),
                 validation_data=val_generator,
                validation_steps=get_steps(len(val_df),batch_size),
                  epochs=10
                 )
```

```python
import matplotlib.pyplot as plot
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'],'b-',label='loss')
plt.plot(history.history['val_loss'],'r--',label='val_loss')
plt.xlabel('Epoch')
plt.legend()
plt.subplot(1,2,2)
plt.plot(history.history['acc'],'b-',label='acc')
plt.plot(history.history['val_acc'],'r--',label='val_acc')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
#테스트데이터로 평가하기
#test_generator: 정답이 없음
model.evaluate(test_generator) #0으로 리턴
```
